```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="species");
```

```python
sns.pairplot(df,hue="species",kind="reg");
```

```python
sns.pairplot(df,kind="reg");
```

```python
import seaborn as sns
flights=sns.load_dataset("flights")
df=flights.copy()
df.head()
```

```python
df1=df.pivot("month","year","passengers")
df1
```

```python
sns.heatmap(df1);
```

```python
sns.heatmap(df1,annot=True, fmt="d");
```

```python
sns.heatmap(df1,annot=True, fmt="d",linewidth=.5);
```

```python
fmri=sns.load_dataset("fmri")
df=fmri.copy()
```

```python
df.head()
```

```python
df.groupby("timepoint")["signal"].count()
```

```python
df.groupby("timepoint")["signal"].mean()
```

```python
df.groupby("subject")["signal"].count()
```

```python
sns.lineplot(x="timepoint",y="signal",data=df);
```

```python
sns.lineplot(x="subject",y="signal",data=df);
```

```python
sns.lineplot(x="timepoint",y="signal",hue="event",data=df);
```
